```python
countrysales = df["Country"].value_counts().reset_index()
countrysales .columns=["Country", "Sales"]
countrysales 
```

```python
name = countrysales["Country"].head(12)
total=countrysales["Sales"].head(12)
sns.set_style("whitegrid")
plt.subplots(figsize=(13,8))

sns.barplot(x= name[0:10],y =total[0:10])
plt.xlabel("Countries",weight="bold")
plt.ylabel("Sales",weight="bold")
plt.grid(alpha = 0.5,axis = "y")
```

```python
Yearsales = df["year"].value_counts().reset_index()
Yearsales .columns=["year", "Sales"]
Yearsales
```

```python
df['year'].value_counts().plot(kind="pie",autopct="%.2f")
plt.show()
```

```python
Monthsales = df["month"].value_counts().reset_index()
Monthsales .columns=["month", "Sales"]
Monthsales
```

```python
name = Monthsales["month"].head(12)
total=Monthsales["Sales"].head(12)
sns.set_style("whitegrid")
plt.subplots(figsize=(13,8))

sns.barplot(x= name[0:10],y =total[0:10])
plt.xlabel("Month",weight="bold")
plt.ylabel("Sales",weight="bold")
plt.grid(alpha = 0.5,axis = "y")
```
